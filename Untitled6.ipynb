```python
#İçeceklerin ve yiyeceklerin besin değerlerini karşılaştırarak hangi kombinasyonun en dengeli olduğunu belirleyebilir miyiz?
```

```python
import pandas as pd

# Veri setini yükleme
df = pd.read_csv("starbucks.csv")

# İçecekler ve yiyecekler arasında dengeli kombinasyonu bulma
def find_balanced_combination(beverages, foods):
    balanced_combinations = []
    for beverage in beverages:
        for food in foods:
            total_calories = beverage['calories'] + food['calories']
            total_fat = beverage['fat'] + food['fat']
            total_carb = beverage['carb'] + food['carb']
            total_fiber = beverage['fiber'] + food['fiber']
            total_protein = beverage['protein'] + food['protein']
            # Örnek olarak, burada bir dengeli kombinasyon kriteri belirlenebilir.
            # Şu anlık sadece kalori miktarını dikkate alıyoruz.
            if total_calories <= 500:
                balanced_combinations.append((beverage['item'], food['item'], total_calories, total_fat, total_carb, total_fiber, total_protein))
    return balanced_combinations

# İçecekler listesi
beverages = [
    {'item': 'Coffee', 'calories': 5, 'fat': 0, 'carb': 0, 'fiber': 0, 'protein': 1},
    {'item': 'Tea', 'calories': 0, 'fat': 0, 'carb': 0, 'fiber': 0, 'protein': 0},
    # İsterseniz veri setinden içecekleri çekebilirsiniz
]

# Yiyecekler listesi
foods = [
    {'item': 'Sandwich', 'calories': 350, 'fat': 15, 'carb': 40, 'fiber': 5, 'protein': 20},
    {'item': 'Salad', 'calories': 200, 'fat': 8, 'carb': 25, 'fiber': 4, 'protein': 15},
    # İsterseniz veri setinden yiyecekleri çekebilirsiniz
]

# Dengeli kombinasyonları bulma
balanced_combinations = find_balanced_combination(beverages, foods)

# Dengeli kombinasyonları yazdırma
for combination in balanced_combinations:
    print(combination)

```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# CSV dosyasını oku
veri_seti = pd.read_csv('starbucks.csv')

# Metin verileri içeren sütunları dışla
numeric_veri_seti = veri_seti.select_dtypes(include=['float64', 'int64']) # Metin verileri içeren sütunlar hariç sayısal sütunları seç

# Min-Max normalizasyonu için MinMaxScaler kullan
scaler = MinMaxScaler()
veri_seti_normal = scaler.fit_transform(numeric_veri_seti)

# Normalleştirilmiş veriyi pandas DataFrame'e dönüştür
veri_seti_normal = pd.DataFrame(veri_seti_normal, columns=numeric_veri_seti.columns)

# Sonucu yazdır
print("Normalleştirilmiş veri seti:")
print(veri_seti_normal)
```

```python
from scipy import stats
import pandas as pd

# Veri setini yükle
veri_seti = pd.read_csv('starbucks.csv')

# 'Kahve' kategorisindeki ürünlerin kalori miktarını al
kahve_kalorileri = veri_seti[veri_seti['type'] == 'kahve']['calories']

# Varsayılan ortalama değer
varsayılan_ortalama = 400

# Z-testini yap
z, p_degeri = stats.ttest_1samp(kahve_kalorileri, varsayılan_ortalama)

# Anlamlılık düzeyini belirle
alfa = 0.05

# H0 hipotezini test et
if p_degeri < alfa:
    print("H0 hipotezi reddedildi: 'Kahve' kategorisindeki ürünlerin ortalama kalori miktarı 400'ten farklıdır.")
else:
    print("H0 hipotezi kabul edildi: 'Kahve' kategorisindeki ürünlerin ortalama kalori miktarı 400'ten farklı değildir.")

# Z istatistiğini ve p-değerini yazdır
print("Z istatistiği:", z)
print("p-değeri:", p_degeri)
```

```python
#'Kahve' kategorisindeki ürünlerin ortalama kalori miktarının 400 olabileceği kabul edilir.
```

```python
import pandas as pd
from scipy.stats import ttest_ind


veri_seti = pd.read_csv('starbucks.csv')

# Bakery ve bistro box kategorilerine ait kalori değerlerini al
bakery_kalorileri = veri_seti[veri_seti['type'] == 'bakery']['calories']
bistro_kalorileri = veri_seti[veri_seti['type'] == 'bistro box']['calories']

# Bağımsız İki Örneklem t-testi uygula
istatistik, p_degeri = ttest_ind(bakery_kalorileri, bistro_kalorileri)

# Sonucu yazdır
print("İstatistiksel Test Sonucu:")
print("t = {:.2f}".format(istatistik))
print("p-değeri = {:.2f}".format(p_degeri))
```

```python
#p-değeri, null hipotezinin doğruluğunu ölçer. Genellikle, p-değeri ne kadar küçükse, null hipotezinin o kadar güçlü bir şekilde reddedildiği kabul edilir.
#Tipik olarak, p-değeri 0.05'ten küçükse, null hipotezi reddedilir ve sonuç istatistiksel olarak anlamlı kabul edilir.
#Ancak, bu durumda p-değeri oldukça yüksek (0.81), bu da null hipotezinin reddedilemeyeceği ve 'bakery' ve 'bistro box' kategorilerine ait kalori değerleri arasında anlamlı bir fark olmadığı anlamına gelir.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV dosyasını oku
veri_seti = pd.read_csv('starbucks.csv')

# Normal dağılımı çizeceğimiz değişkeni seç
secilen_degisken = 'calories'

# Histogram ve yoğunluk eğrisini çiz
plt.figure(figsize=(8, 6))
sns.histplot(veri_seti[secilen_degisken], kde=True, color='blue', bins=30)
plt.title('Normal Dağılım Grafiği (' + secilen_degisken + ' Değişkeni)')
plt.xlabel('Değerler')
plt.ylabel('Yoğunluk')
plt.grid(True)
plt.show()
```

```python
from scipy.stats import binomtest

# Örnek veri seti
df = pd.read_csv("starbucks.csv")

# Ürün kategorisi ve belirli bir değer
kategori = "bakery"
belirli_deger = 350  # Test etmek istediğimiz belirli bir değer

# Belirli ürün kategorisine ait kalori değerleri
kategori_kalorileri = df[df["type"] == kategori]["calories"]

# Tek anakütle oran testi
p_value = binomtest(sum(kategori_kalorileri > belirli_deger), len(kategori_kalorileri), p=0.5)

print("p-değeri:", p_value)
```

```python
# tek anakütle oran testini hesapladık  41 denemeden 24 başarılı olduğunu söylüyor. istatistik değerini ve p değerini gösteriyor.
```

```python
import pandas as pd
from scipy.stats import ttest_ind
df = pd.read_csv("starbucks.csv")

# Örnek olarak iki grup oluşturalım (örneğin, 'bakery' ve 'sandwich' türlerindeki ürünler)
bakery = df[df['type'] == 'bakery']['calories']
sandwich = df[df['type'] == 'sandwich']['calories']

# T testini uygulayalım
t_statistic, p_value = ttest_ind(bakery, sandwich)

# Sonucu yazdıralım
print("T test istatistiği:", t_statistic)
print("P değeri:", p_value)  #fırın ürüneri ve sandviç ürünleri arasında t testi hesapladım
```
